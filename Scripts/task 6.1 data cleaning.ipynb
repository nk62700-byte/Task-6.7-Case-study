```python
# Importing libraries

import pandas as pd
```

```python
# creating path

path = r'C:\Users\nikik\achievement 6\boat_data.csv'
```

```python
# imoirting datafile
df = pd.read_csv(path)
```

```python
df.head()
```

```python
#Renaming columns
df.columns = [col.replace('\n', ' ').strip() for col in df.columns]

df.rename(columns={'Year Built': 'Year', 'Number of views last 7 days': 'Number of views'}, inplace = True)
```

```python
df.columns
```

```python
# Missing values
df.isna().sum()
```

```python
# filling the missinhg values using median
df["Length"].fillna(df["Length"].median(), inplace=True)
df["Width"].fillna(df["Width"].median(), inplace=True)
```

```python
# replacing missing values with unknown
df["Manufacturer"].fillna("Unknown", inplace=True)
df["Material"].fillna("Unknown", inplace=True)
```

```python
# checking for duplicate 
df.duplicated().sum()
```

# No duplicates were found

```python
# Remove currency symbols and text from the Price column
df['price_clean'] = (df['Price'] .str.replace(r"[A-Z]{3}", "", regex=True).str.strip())
```

```python
#Convert cleaned price column to numeric
# Invalid values will be converted to NaN
df['price_clean'] = pd.to_numeric(df['price_clean'], errors='coerce')
```

```python
df[['Price', 'price_clean']].head()
```

```python
df.describe()
```

```python
# Define file paths for cleaned data
cleaned_data = r'C:\Users\nikik\achievement 6\boat_data_cleaned.csv'
df.to_csv(cleaned_data, index = False)
```
